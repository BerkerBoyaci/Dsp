## 1.A-)


İşaretlerin 2200Hz ile örneklenmesinden sonra; 

x1[n] = 7 cos( pi * n / 11 )
x2[n] = 3 cos( pi * n / 11 ) + 4 cos( 9 * pi * n / 11) 
olarak bulunur.

w = ( pi  ) / 11 için N = 22
w = ( 9 * pi ) / 11 için N = 22 hesaplanır.



7 cos( 200𝜋𝑡 ) ve 3 cos( 200𝜋𝑡 ) için frekans 100Hz , 4cos⁡(1800𝜋𝑡) için frekans 900Hz olur.Fn için sistemin en büyük frekansı 900Hz seçilir. Fs = 2200Hz , 2Fn = 1800Hz için 𝐹𝑠 ≥ 2 𝐹N şartını sağladığından  örtüşme olmaz. Örtüşme olmadığından dolayı iki sinyal arasında bir benzerlik bulunmaz.

```python
Fs = 2200
Ts = 1/Fs

N = 22
N1 = 22
N2 = 22
N3 = 22

n  = np.arange( 0 , 2 * N )
n1 = np.arange( 0 , 2 * N1 )
n2 = np.arange( 0 , 2 * N2 )
n3 = np.arange( 0 , 2 * N3 )

x1 = 7 * np.cos( 200 * np.pi * n1 * Ts )
x2 = 3 * np.cos( 200 * np.pi * n2 * Ts ) + 4 * np.cos( 1800 * np.pi * n3 * Ts)

plt.figure()
plt.stem( n , x1 , use_line_collection= 'True')
plt.title('x[n]')
plt.xlabel('örnek')
plt.ylabel('x[n]')
plt.show()

plt.figure()
plt.stem( n , x2 , use_line_collection= 'True')
plt.title('x[n]')
plt.xlabel('örnek')
plt.ylabel('x[n]')
plt.show()
```

## B-)

```python
def freqPlot(xn , N ):
    w = np.arange(-np.pi, np.pi, 2*np.pi/N) 
    xw = np.fft.fftshift(np.fft.fft(xn,N)/N)
    
    plt.figure()
    plt.subplot(1,2,1)
    plt.stem( w / np.pi , np.abs(xw) ,use_line_collection='True')
    plt.xlabel("$\omega$ / $\pi$") 
    plt.title("$X(\omega)$ (Genlik)")

    plt.subplot(1,2,2)
    plt.stem( w / np.pi , np.angle(xw) ,use_line_collection='True')
    plt.xlabel("$\omega$ / $\pi$") 
    plt.title("$X(\omega)$ (Faz)")
    plt.show()
    
```

```python
freqPlot( x1 , N )
freqPlot( x2 , N )
```

```python
def quantize(x, S):
    """
    Kuantalama fonksiyonu
    """
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
 
    dists = abs(X-S)
    

    
    nearestIndex = dists.argmin(axis=1)
    
    quantized = S.flat[nearestIndex]
        
    print(quantized)
    print('---------------')
    
    return quantized.reshape(x.shape)
```

```python
def getQuantize(xn , k_bit):
    """
    Kuantalama fonksiyonuna gönderilecek dizinin hesaplanması ve
    fonksiyonun gönderdiği değerleri iletme için
    
    Giriş parametreleri :
    xn    : quantize fonksiyonuna gönderilecek numpy array'i
    k_bit : bit sayısı
    
    Çıkış ifadeleri :
    y_midtrise : mid trise yöntemi ile
    y_midtread : mid thread yöntemi ile
    
    """
    
    Max_range= np.amax(xn)-np.amin(xn)
    q=pow(2,k_bit)
    delta = Max_range/q
    S_midrise = -np.amax(xn)  + np.arange(q)*delta + delta/2
    S_midtread = -np.amax(xn) + np.arange(q)*delta


    #y_midtrise = quantize(xn, S_midrise) 
    #y_midtread = quantize(xn, S_midtread)
    
    return (quantize(xn, S_midrise)  , quantize(xn, S_midtread) )
```

```python

𝑥n = np.array( [ 1.25 , 3.5 ,4 , 6.7 , -1.4 , -3.5 , -10.5 , 8.5 , 3.8 ] )
n = np.arange(len(𝑥n))


k_bit = 3
y_midtrise , y_midtread  = getQuantize(xn , k_bit)



plt.figure() 
plt.stem( n , xn , use_line_collection = 'True')
plt.bar(  n , y_midtrise , color='orange' )
plt.xlabel("örnek") ; 
plt.title('midrise yöntemiyle kuantalanmış işaret')

plt.figure()
plt.stem( n , xn , use_line_collection = 'True')
plt.bar(  n , y_midtread , color='orange')
plt.xlabel("örnek") ; 
plt.title('midtread yöntemiyle kuantalanmış işaret')
plt.show()
```
