```python
Fs = 2200
Ts = 1/Fs

N = 22
N1 = 22
N2 = 22
N3 = 22

n  = np.arange( 0 , 2 * N )
n1 = np.arange( 0 , 2 * N1 )
n2 = np.arange( 0 , 2 * N2 )
n3 = np.arange( 0 , 2 * N3 )

x1 = 7 * np.cos( 200 * np.pi * n1 * Ts )
x2 = 3 * np.cos( 200 * np.pi * n2 * Ts ) + 4 * np.cos( 1800 * np.pi * n3 * Ts)

plt.figure()
plt.stem( n , x1 , use_line_collection= 'True')
plt.title('x[n]')
plt.xlabel('√∂rnek')
plt.ylabel('x[n]')
plt.show()

plt.figure()
plt.stem( n , x2 , use_line_collection= 'True')
plt.title('x[n]')
plt.xlabel('√∂rnek')
plt.ylabel('x[n]')
plt.show()
```

```python
def freqPlot(xn , N ):
    w = np.arange(-np.pi, np.pi, 2*np.pi/N) 
    xw = np.fft.fftshift(np.fft.fft(xn,N)/N)
    
    plt.figure()
    plt.subplot(1,2,1)
    plt.stem( w / np.pi , np.abs(xw) ,use_line_collection='True')
    plt.xlabel("$\omega$ / $\pi$") 
    plt.title("$X(\omega)$ (Genlik)")

    plt.subplot(1,2,2)
    plt.stem( w / np.pi , np.angle(xw) ,use_line_collection='True')
    plt.xlabel("$\omega$ / $\pi$") 
    plt.title("$X(\omega)$ (Faz)")
    plt.show()
    
```

```python
freqPlot( x1 , N )
freqPlot( x2 , N )
```

```python
def quantize(x, S):
    """
    Kuantalama fonksiyonu
    """
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
 
    dists = abs(X-S)
    

    
    nearestIndex = dists.argmin(axis=1)
    
    quantized = S.flat[nearestIndex]
        
    print(quantized)
    print('---------------')
    
    return quantized.reshape(x.shape)
```

```python
def getQuantize(xn , k_bit):
    """
    Kuantalama fonksiyonuna g√∂nderilecek dizinin hesaplanmasƒ± ve
    fonksiyonun g√∂nderdiƒüi deƒüerleri iletme i√ßin
    
    Giri≈ü parametreleri :
    xn    : quantize fonksiyonuna g√∂nderilecek numpy array'i
    k_bit : bit sayƒ±sƒ±
    
    √áƒ±kƒ±≈ü ifadeleri :
    y_midtrise : mid trise y√∂ntemi ile
    y_midtread : mid thread y√∂ntemi ile
    
    """
    
    Max_range= np.amax(xn)-np.amin(xn)
    q=pow(2,k_bit)
    delta = Max_range/q
    S_midrise = -np.amax(xn)  + np.arange(q)*delta + delta/2
    S_midtread = -np.amax(xn) + np.arange(q)*delta


    #y_midtrise = quantize(xn, S_midrise) 
    #y_midtread = quantize(xn, S_midtread)
    
    return (quantize(xn, S_midrise)  , quantize(xn, S_midtread) )
```
