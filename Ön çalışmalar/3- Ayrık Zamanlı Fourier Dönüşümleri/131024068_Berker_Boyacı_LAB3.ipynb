```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, ifft , fftshift
%matplotlib notebook
```

## Soru - 1 

```python
w = np.linspace( 0 , 2*np.pi ,1000 )
H = 1 - np.e**(-1j * w)

#N = len(H)
H_abs = np.abs(H )
H_phase = np.angle(H )

plt.figure()
plt.subplot(121)
plt.plot(w / np.pi , H_abs)
plt.grid()
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}|X(e^{j\omega})|$')

plt.subplot(122)
plt.plot(w / np.pi , H_phase)
plt.grid()
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}\measuredangle X(e^{j\omega})(\\times\pi)$')
```

## Soru 1-b :

```python
w = np.linspace( -1*np.pi , np.pi ,1000 )
H = 1 - np.e**(-1j * w)


H_shift = fftshift(H)
H_abs = np.abs(H_shift )
H_phase = np.angle(H_shift )

plt.figure()
plt.subplot(121)
plt.plot(w / np.pi , H_abs)
plt.grid()
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}|X(e^{j\omega})|$')

plt.subplot(122)
plt.plot(w / np.pi , H_phase)
plt.grid()
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}\measuredangle X(e^{j\omega})(\\times\pi)$')
```

Grafikleri yorumlayabilmek için A şıkkında 0-2pi yerine -pi~pi aralığında gösterilip H değeri fftshift ile kaydırılırsa Alçak-geçiren filtre karakteristiği gösterdiği söylenebilir.

###### Soru 1 - c :

```python
n = np.arange(0,16)
h_n = np.append(np.ones(2),np.zeros(14))
N = len(h_n)

w_disc = n * 2 * np.pi / N

X_fft_shift = fftshift(fft(h_n , 16))
X_abs   = np.abs  ( X_fft_shift )
X_phase = np.angle( X_fft_shift )

plt.figure()
plt.subplot(121)
plt.stem(w_disc / np.pi , X_abs ,use_line_collection = 'True')
plt.grid()
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}|X(e^{j\omega})|$')

plt.subplot(122)
plt.stem(w_disc / np.pi , X_phase/np.pi , use_line_collection = 'True')
plt.grid()
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}\measuredangle X(e^{j\omega})(\\times\pi)$')
```
