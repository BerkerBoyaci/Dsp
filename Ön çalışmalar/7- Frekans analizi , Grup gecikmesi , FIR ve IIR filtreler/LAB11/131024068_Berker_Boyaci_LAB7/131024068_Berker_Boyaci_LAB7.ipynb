```python
import numpy as np             
import matplotlib.pyplot as plt
import scipy.signal as sgnl
from scipy.fftpack import fft, ifft
import zplane

import SystemFunction
from scipy.io import loadmat

%matplotlib notebook
```

# Soru 1-)

###### Type 1 filtre ->

```python
N = 100
hn = np.ones(5)
Hw = fft(hn, N)

w = np.linspace( 0 , 2 * np.pi, N)

Hw_mag = abs(Hw)
Hw_phs = np.unwrap(np.angle(Hw))

tau_w = -np.diff(Hw_phs)*(N/(2*np.pi))


plt.figure()
plt.plot(w/np.pi, Hw_mag)
plt.xlabel('frequency (rad/sample)')
plt.ylabel('magnitude response)')
plt.title('magnitude response')

plt.figure()
plt.plot(w/np.pi, Hw_phs)
plt.ylabel('phase (rad)')
plt.xlabel('frequency rad/sample')

plt.figure()
plt.plot(w[1::]/np.pi, tau_w)
plt.xlabel('frequency (rad/sample)')
plt.ylabel('group delay (samples)')
```

Genlik cevabına bakıldığında 0'da en yüksek pi'de en düşük değeri aldığından dolayı alçak geçiren filtre özelliği gösterir. <br><br>
tip 1 filtre olduğu için alçak geçiren , yüksek geçiren , bant geçiren ve bant durduran filtreler olabilirdi. Genlik grafiğinden gözlenen sonuca uygun. <br><br>
Geçirme bandında lineer faza ve sabit grup gecikmesine sahip olduğu grafiklerden görülür.

```python
hn = np.ones(5)

w , h = sgnl.freqz(hn)
Hw, polezz, zerozz, K = SystemFunction.SystemFunction(w)

num , denum = sgnl.zpk2tf(zerozz , polezz , K)


zplane.zplane(num , denum)
```
