## Soru-1 : 

a = 3 + 4j , b = 1 - 2j için c = a + b , d = a / b hesapla. c ve d'nin gerçek , sanal , genlik ve faz değerlerini bastır

```python
a = 3 + 4j
b = 1 - 2j

c = a + b
d = a / b
```

```python
print("c'nin genliği       : ",np.abs(c)   , "\n",
      "c'nin fazı          : ",np.angle(c) , "\n",
      "c'nin gerçek değeri : ",np.real(c)  , "\n",
      "c'nin Sanal değeri  : ",np.imag(c)  , "\n")
```

```python
print("d'nin genliği       : ",np.abs(d)   , "\n",
      "d'nin fazı          : ",np.angle(d) , "\n",
      "d'nin gerçek değeri : ",np.real(d)  , "\n",
      "d'nin Sanal değeri  : ",np.imag(d)  , "\n")
```

## Soru-2 : 

    x[n] = ⁡2cos⁡(2 pi n / 7 + pi / 5 ) işaretini 2 periyot bastırmak için 14'e kadar aralık belirledik. periyodu elle 7 bulundu.

```python
n = np.arange(0,14)
x_n = 2 * np.cos(2 * np.pi * n / 7 + np.pi / 5 ) 
```

```python
plt.stem(n, x_n , use_line_collection = "True")
plt.xlabel(" n ")
plt.ylabel(" x[n] ")
plt.title( " x[n] ")

plt.annotate("1. Period \n başlangıcı", xy = (0,1.5)
             ,xytext =(0.2 , 0.7)
             ,arrowprops=dict(facecolor ='green',shrink = 0.05),)

plt.annotate("2. Period \n başlangıcı", xy = (7,1.5)
             ,xytext =(7.2 , 0.7)
             ,arrowprops=dict(facecolor ='red',shrink = 0.05),)
plt.show()
```

```python
n = np.arange(0,10)
```

 A - ) exp(j pi n / 4 )

```python
x1_n = np.e**( ( 1j * np.pi *n ) / 4) 

plt.stem( n , np.real(x1_n) , use_line_collection = "True")
plt.xlabel( " n " )
plt.ylabel("x1_n " )
plt.title(" x1_n reel")
plt.show()

plt.figure()


plt.stem( n , np.real(x1_n) , use_line_collection = "True")
plt.xlabel( " n " )
plt.ylabel("x1_n " )
plt.title(" x1_n imajiner ")
plt.show()
```

```python
x2_n = np.e**( ( 1j * np.pi ) / 4) + n - n

plt.stem( n , np.real(x2_n) , use_line_collection = "True")
plt.xlabel( " n " )
plt.ylabel("x2_n " )
plt.title(" x2_n reel")
plt.show()

plt.figure()


plt.stem( n , np.real(x2_n) , use_line_collection = "True")
plt.xlabel( " n " )
plt.ylabel("x2_n " )
plt.title(" x2_n imajiner ")
plt.show()
```
